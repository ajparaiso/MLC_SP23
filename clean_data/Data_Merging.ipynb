### 311 Data

```python
# Read

calls = pd.read_csv('./MLC_SP23/raw_data/calls_311_data.csv', index_col=[0])
calls
```

```python
calls.isnull().values.any()
```

```python
calls_col = ['BBL'] + ['HPD_calls_' + str(i) for i in range(2017,2020)] + ['DOB_calls_' + str(i) for i in range(2017,2020)]
calls_col
```
