```python
# Import Data

import pandas as pd
import numpy as np

file_path_1 = '/Users/andrewparaiso/Documents/GitHub/MLC_SP23/clean_data/merged_data_v3.csv'
file_path_2 = './MLC_SP23/clean_data/merged_data_v3.csv'

df = pd.read_csv(file_path_1, low_memory=False, index_col=[0])

X = df.drop(['rs_share_diff_new'], axis = 1)
Y = df['rs_share_diff_new']
```

```python
df.head()
```

```python
# Standardize Data
from sklearn import preprocessing

X_scaled = preprocessing.scale(X)
X = X_scaled

# Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=1)
```

```python
# Baseline accuracy
1 - Y_train.mean()
```

```python
# Fit SVM
from sklearn import svm
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
import time

scoring = {'Accuracy':'accuracy',
           'F1 Score':'f1',
           'Precision':'precision',
           'Recall':'recall',
           'AUC':'roc_auc'
}

param_grid = {'C': [0.1, 1, 10],
              'kernel': ['linear']
}

start = time.time()

svm_clf = svm.SVC(probability = True, random_state = 1)
gr = GridSearchCV(svm_clf, param_grid = param_grid, scoring = scoring, refit = "AUC", cv = 4)
fit = gr.fit(X_train, Y_train)

end = time.time()

print("Running grid search took", end - start, "seconds") # time in seconds to run grid search
print()
print(fit.best_params_)
print(fit.best_score_)
```
